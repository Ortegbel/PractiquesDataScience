# Es hora de juntar els datasets



```python
movies_rating = (ratings.set_index("movie_id").join(movies.set_index("movie_id"),how="left"))
movies_rating
```

Per separar el generes he descovert la funció get_dumies, q ens retornarà un 0 ó un 1  a la casella corresponent.

```python
dummies = movies_rating['genres'].str.get_dummies()
dummies.head()
```

```python
#i ara concatenem els dataframes.
tidy_movie_ratings = (pd.concat([movies_rating, dummies], axis=1).drop(["rating_timestamp", "genres"], axis=1) )

tidy_movie_ratings.head()
```

```python
#afegim l'any de produccio en una columna separada, i l'eliminem del nom. Tal com haviem fet en la pràctica anterior.
tidy_movie_ratings["production_year"] = tidy_movie_ratings["movie_title"].str[-5:-1]
tidy_movie_ratings["movie_title"] = tidy_movie_ratings["movie_title"].str[:-7]
```

```python
tidy_movie_ratings.reset_index(inplace=True)
tidy_movie_ratings.head(2)
#i ja ho tenim, llest per començar.
```

# visualitzacions


En aquest apartat hi han parts de codi que segueixen 1:1 l'exemple mencionat abans.
Són exemples que em serveixen de referència personal per tenir en compte i aprendre.
Tot i així, els intento replicar d'una maner diferent.


```python
# L'he pres de referència per entendre com aplicar condicions quan agrupem.
""" 
First, you will start with visualizing the total volume of films created over the years.

Next, you will count the total number of productions for each year and plot it. 
The record you see for the year of 2021 should be filtered out before proceeding:

"""

condition = tidy_movie_ratings["production_year"].astype(int) < 2021

prodcount = (tidy_movie_ratings[condition][["production_year", "movie_id"]].groupby("production_year").count())

prodcount.tail()
```

```python
#You can chart a 5 year moving average of the total productions:

(prodcount.rolling(5).mean().rename(columns={"movie_id":"count"}).plot(figsize=(15,5),title="Count of Rated Movies - by production year"))
```

#### A mi no em donaba el mateix resultat i he porbat de fer-ho diferent.

```python

prodcount2 = (tidy_movie_ratings[condition][["production_year", "movie_id"]].groupby("production_year")["movie_id"].count())
prodcount2.reset_index()
```

```python
plt.figure(figsize=(10,6))
ax= sns.lineplot( data= prodcount2)
```

ha quedat molt semblant.

#### Una questió que m'interessava comparar 
m'interessava extreure'n quin tipus de pel.licules eren més populars durant la Segona Guerra Mundial, serie, les d'acciò, les de drama...

```python
genre_groups_WWII = (tidy_movie_ratings.iloc[:, 4:]
                .groupby("production_year")
                .sum()
               ).loc["1940":"1945"] 

```

```python
genre_groups_WWII=genre_groups_WWII.loc[:, (genre_groups_WWII != 0).any()]
genre_groups_WWII
```

Però no aconsegueixo que quedi net, i aquesta és la millor representació que n'he obtingut.
