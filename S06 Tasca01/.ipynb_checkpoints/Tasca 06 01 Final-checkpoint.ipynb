## SMOTE 
Aquest pas ha estat el més complicat per no triar les dades correctes per fer la mostra.

SMOTE és l'acrònim de Synthetic Minority Oversampling Technique, i ens permet crear mostres sintètiques per poder equilibrar la representació dins la mostra de variables poc representades.

Té diferents variants, dependen de si totes les variables son numèriques SMOTE, totes son categòriques SMOTEN o apareixen dades dels dos tipus SMOTENC.

En un principi vaig triar la variable SMOTEN, però em retornava errors estranys, com si hi hagués una fila de dades que no li permetés fer l'operació, finalment vaig decidir factoritzar, netejar les dades i  començar denou i aquí tenim el resultat.

Fent un .head() i un .info(), veig millor de quines dades disposem.
Al primer cop d'ull veiem que tenim molts caps nuls NaN i algunes columnes que contenen objectes, dues coses a solucionar:

```python
df.head()
```

```python

df.info()
```

amb el següent codi, eliminaré les columnes que no necessito a més de canviar per 0 els valors nuls.

```python
df= df.fillna(0)
df = df.drop (['confID','divID','franchID','name'], axis=1)
df.info()
```

Per tal de poder fer servir la columna __playoff__ reviso les dades que conté, i faig les passes necessàries per convertir les seves dades en 0 pels quips que no van accedir al play-off, i 1 pels que sí que ho van fer:

```python
df.playoff.unique()
```

```python
import re

df['playoff'] = df.playoff.replace(regex='\D', value=1)
df.playoff.unique()
```

per últim, factoritzem els camps que pertanyen a la identificació de la lliga __lgID__ i de l'equip __tmID__:

```python
df['lgID'] = pd.factorize(df['lgID'])[0]
df['tmID'] = pd.factorize(df['tmID'])[0]
df.head()
```

Ara que ja ho tinc tot en valors numèrics, importaré les llibreries necessàries i aplicaré el mètode SMOTE:

```python
import imblearn
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy = 'minority')
X_sm, y_sm = smote.fit_resample(df, df['playoff'])
```

Fent un .value_counts() de la variable original i de la mostra SMOTE comprovem que sen's han generat les mostres sintètiques a la variable minoritària.

```python
df.playoff.value_counts()
```

```python
X_sm.playoff.value_counts()
```

##  Reservoir Sampling

El __Reservoir Sampling__ ens és útil quan tenim un riu de dades (Stream) de llargada desconeguda i tan gran que només el podem iterar un cop per fer la mostra.  
Aquest algorisme ens permet crear un reservori on guardarem els elements que triem aleatòriament en una primera volta, i després, d'aquests elements reservats, en triarà aleatòriament el número d'entrades que haguem definit prèviament.



Aquesta solució, es basa a l'article abans mencionat de Rahul Agarwal.

```python
import random

# com stream prenem la columna que conté els anys
stream = df['year']
# fent el el reservori de mostres de l'stream
k=50
reservoir = []
for i, element in enumerate(stream):
    if i+1<= k:
        reservoir.append(element)
    else:
        probability = k/(i+1)
        if random.random() < probability:
            # Selecionem aleatoriament un dels elements i treiem un dels elements que ja havia estat seleccionat.
             reservoir[random.choice(range(0,k))] = element

print(reservoir)
```
