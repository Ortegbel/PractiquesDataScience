Exercici 1 Descarrega el data set Airlines Delay: Airline on-time statistics and delay causes i carrega’l a un pandas Dataframe. Explora les dades que conté, i queda’t únicament amb les columnes que consideris rellevants.

```python
import pandas as pd
"""
El primer q hem de fer és importar la llibreria de Pandas, q ho farem amb el seu àlies pd, i obrir el document q hem descarregat amb la funció 
pd.read_"extensiodel document"("nom i ubicació del document")
Com es tracta d'un document tant pesat l'he comprimit en gzip i ara, en llegir-ho, haig d'especificar quina compressió he emprat.
He afegit com a parametre la columna index = Unique carrier, per tenir -ho ordenat per companyia

per veure les dades q conté llistaré les primeres columnes amb la funció head()

"""
dades = pd.read_csv("DelayedFlightsComp.csv.gz", compression='gzip', index_col= "UniqueCarrier")
dades.head()
```

```python
"""
He afegit i manipulat les columnes necessàries al DataFrame original, així després puc triar només les columnes q més em convinguin.

He convertit les dades en milles i minuts a Km i hores per poder calcular la velocitat mitjana.
Per tal de poder agrupar després el vol per origen i destinació he creat una columna amb aquestes dades i per això he hagut de convertir el numero de vol en str.
"""
dades['fDetails'] = dades['Origin'] + ' '+ dades['Dest']
dades['Distance KM'] = (dades['Distance'] * 1.609)
dades['AirTime h'] = (dades["AirTime"]/60)
dades['MeanSpeed Km/h'] = (dades['Distance KM'] / dades ['AirTime h'])
dades["FlightNum2"] = dades["FlightNum"].astype(str)
dades['FlightDetails']= dades["FlightNum2"] + ' '+ dades['fDetails']

dades_1 = dades[["DepTime","CRSDepTime","ArrTime","CRSArrTime","AirTime","ArrDelay","DepDelay","Distance", 'MeanSpeed Km/h',"FlightNum", "Origin","Dest","FlightDetails"]]


dades_1.head()
```

Exercici 2 Fes un informe complet del data set:.
Resumeix estadísticament les columnes d’interès Troba quantes dades faltants hi ha per columna Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...) Taula de les aerolínies amb més endarreriments acumulats Quins són els vols més llargs? I els més endarrerits? Etc.

```python
"""
Amb el metode info() veurem quantes dades tenim finalment i de quin tipus són.
"""

dades_1.info()
```

```python
"""
Pel tipus de dades que estem treballant i la baixa incidència de les dades nul.les, crec q serà millor eliminar-les

Com el DataFrame té més d'un mi.lio de regristres i les dades nul.les representa on 0.4% i un 0.01%.
"""
dades_1= dades_1.dropna()
```

```python
dades_1.isnull().sum()
```

```python
"""
Per tal d'afegir una columna amb el resultat de dues condicions he optat per aquesta solució.
He considerat q el vol compta com endarrerit, tot i que es recuperi el temps inicial d'espera durant el vol, ja que ha sortit
amb retard.
"""
def myfunc (ArrDelay, DepDelay ):
    if ArrDelay >0  :
        TotalDelay =  1
    elif DepDelay>0:
        TotalDelay = 1
    else:
        TotalDelay = 0
    return TotalDelay
        
dades_1['TotalDelay'] = dades_1.apply(lambda x: myfunc (x['ArrDelay'], x['DepDelay']), axis=1)
dades_1.head()
```

```python
"""
Per últim, afegirem el nom de la companyia q opera el vol.
Per això he baixat el fitxer amb la correspondencia entre el distintiu i el nom complert.

"""
carriers= pd.read_csv('carriers.csv', index_col= "UniqueCarrier")
carriers.head()
```

```python
dades_2=dades_1.merge(carriers, left_index=True, right_index= True)
dades_2.head()
```

```python
"""
Les aerolinees amb més endarreriments són
"""
dadesAgrupades = dades_2.groupby(['Description', 'UniqueCarrier'], sort=False)[['ArrDelay']].mean()
dadesAgrupades.sort_index(ascending= False).head(5)
```

```python
"""
els vols més llargs
"""

dadesVols= dades_1.groupby(['FlightDetails'], sort=False)[['Distance']].max()
dadesVols.sort_values(['Distance'], ascending=False).head(5)
```

```python
"""
els més endarrerits
"""
VolsEndarrerits = dades_1.groupby(['FlightDetails'], sort=False)[['ArrDelay']].max()
VolsEndarrerits.sort_values(['ArrDelay'],ascending= False).head(5)
```
