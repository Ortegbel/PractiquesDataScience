# Agafa un text en anglès que vulguis, i calcula'n la freqüència de les paraules Treu les stopwords i realitza stemming al teu conjunt de dades.

He triat un text disponible al corpus de nltk, el discurs de l'estat de la nació de Kennedy de l'any 61 en versió Raw, per poder fer tots els passos. 

```python
from nltk.corpus import state_union
text= state_union.raw('1961-Kennedy.txt')
```

Per poder calcular la freqüència de les paraules, dividirem el text en paraules i n'eliminarem els stopwords.

```python

from nltk.tokenize import TreebankWordTokenizer
tokenizer_wrd = TreebankWordTokenizer()
words=tokenizer_wrd.tokenize(text)


from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))

revisedText= [word for word in words if (word not in stop_words)]


from nltk import FreqDist
frequency_distribution = FreqDist(revisedText)
print(frequency_distribution)
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt
frequency_distribution.plot(30,cumulative=False)
plt.show()
```

Aquests serien els reusltats de no haver eliminat els stopwords.

```python
frequency_distribution = FreqDist(words)
print(frequency_distribution)
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt
frequency_distribution.plot(30,cumulative=False)
plt.show()
```

## Stop Words i Stemming

He trobat un altre text per fer aquesta part de l'exercici, es tracta del monòleg final del replicant Roy Batty a Balder Runner.


```python
import nltk
from nltk.corpus import stopwords
```

```python
#blader runner
text = open('bladeRunner.txt', 'r',encoding="utf-8").read()
text 
```

He probat de fer un RegexpTokenizer, per fer un tokenize i eliminar els signes de puntuació.

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r"\w+")
text3 = tokenizer.tokenize(text) 
```

```python
#treure els Stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))

revisedText= [word for word in text3 if (word not in stop_words)]
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
stemm_text = [ps.stem(words)for words in revisedText]
stemm_text
```

# Realitza sentiment analysis al teu conjunt de dades.

per fer el Sentimen Analysis, he trobat el següent grup de dades amb les valoracions de la pàgina IMDB.

```python
import pandas as pd 

df = pd.read_csv('IMDB Dataset.csv')
df.shape
```

```python
df.head()
```

```python
df.sentiment.value_counts()
```

S'han detectat etiquetes de salt de linea que s'eliminaran amb el següent codi

```python
df['review'] = df['review'].str.replace('<br />', '')
df.head()
```

El sentiment analysis es fa amb Vader i la funció SentimentIntensityAnalyzer

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# amb Vader i la funció SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
```

```python
df['neg'] = df['review'].apply(lambda x:sia.polarity_scores(x)['neg'])
df['neu'] = df['review'].apply(lambda x:sia.polarity_scores(x)['neu'])
df['pos'] = df['review'].apply(lambda x:sia.polarity_scores(x)['pos'])
df['compound'] = df['review'].apply(lambda x:sia.polarity_scores(x)['compound'])

df.head()
```

```python
vader_comp = df["compound"].mean()

print("La polaritat mitjana de les comentaris de les pel.licules es:",vader_comp)
```
