**Exercici 2:**

l'anàlisi de components principals: **PCA**.  




```python
import pandas as pd 

dades2 = pd.read_csv('dades/Master.csv',usecols=['firstName', 'lastName', 'pos', 'height', 'weight','shootCatch'])
dades2['name'] = dades2['firstName']+ dades2['lastName']
dades2 = pd.DataFrame(dades2[['name','pos','height','weight','shootCatch']])
dades2.head()
```

```python
dades2.pos.unique()
```

```python
#preparem la variable pos per distingir entre porters i no porters.

dades2.loc[(dades2.pos != 'G'), 'pos'] = 'P'
dades2.head()
```

```python
dades2 = pd.get_dummies(data = dades2, columns=['shootCatch'])
dades2.head()
```

```python
dades2 = dades2.dropna()
```

```python
dades2.shape
```

```python
#comencem
from sklearn.preprocessing import StandardScaler
```

```python
#separem les columnes caracteristiques de l'objectiu

caract = ['height','weight','shootCatch_B','shootCatch_L','shootCatch_R']
x = dades2.loc[:, caract].values

y = dades2.loc[:,['pos']].values
```

```python
# Standarditzo les caracteristiques
x = StandardScaler().fit_transform(x)
```

```python
pd.DataFrame(data = x, columns = caract).head()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
pca = PCA(n_components=2)
```

```python
principalComponents = pca.fit_transform(x)
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf.head()
```

```python
#ara concatenem les dades

finalDf = pd.concat([principalDf, dades2[['pos']]], axis = 1)
finalDf.head(5)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


position = ['P','G']
colors = ['r', 'g']

for position, color in zip(position,colors):
    indicesToKeep = finalDf['pos'] == position
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)

ax.legend(position)
ax.grid()
```
