### Exercici 1 

Resumeix gràficament el data set DelayedFlights.csv. 

Crea almenys una visualització per: 
Una variable categòrica (UniqueCarrier)
Una variable numèrica (ArrDelay)
Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
Dues variables numèriques (ArrDelay i DepDelay)
Tres variables (ArrDelay, DepDelay i UniqueCarrier)
Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).

```python
#com sempre, primer importarem les llibreries q emprarem així com el DataFrame amb q treballarem.

%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('DelayedFlightsComp.csv.gz', compression ='gzip')
df.head()
```

```python
#n'extreure'm les dades q necessitarem i les guardarem a un DF més petit.

dades = df[["UniqueCarrier", "ArrDelay","DepDelay","Distance","FlightNum", "AirTime"]]


dades.head()
```

Amb aquest nou DataFrame comenarem a fer les gràfiques i les guardarem com imatges.

Una variable categòrica (UniqueCarrier)

```python
#generaré el gràfic i el guardadaré com una imatge .png
ax = sns.stripplot(x='UniqueCarrier', data = dades)
plt.savefig('imatge_1.png')
```

Una variable numèrica (ArrDelay)

```python

x = dades['ArrDelay'].values


sns.displot(x)

plt.savefig('imatge_2.png')
```

Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)

```python
ax = sns.barplot(x = 'UniqueCarrier', y = 'ArrDelay', data = dades,)
plt.savefig('imatge_3.png')
```

Dues variables numèriques (ArrDelay i DepDelay)

```python
ax = sns.relplot(x = 'ArrDelay', y = 'DepDelay',data = dades)
plt.savefig('imatge_4.png')
```

Tres variables (ArrDelay, DepDelay i UniqueCarrier)

```python
ax = sns.lmplot(x = 'ArrDelay', y = 'DepDelay', hue = 'UniqueCarrier',data = df)
plt.savefig('imatge_5.png')
```

Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier)

```python

ax = sns.scatterplot(data = dades, x= 'UniqueCarrier', y= 'AirTime', hue='ArrDelay', style= 'DepDelay', legend= False)
plt.savefig('imatge_6.png')
```
