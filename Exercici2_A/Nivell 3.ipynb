Utilitzar el mètode, mpimg.imsave () de la llibreria importada, per guardar les imatges modificades i que haureu de pujar al vostre repositori a github.

Eliminaré un dels colors, convertint el numpy array en dos dimensions enlloc de tres. no importa quin eliminem, el resultat és el mateix ja q matplotlib afegeix un colormap de manera automàtica.

```python
imatge= plt.imread ('img3.jpg')
imatge0= imatge[:, :, 0]
plt.imshow(imatge0)
```

```python
imatge= plt.imread ('img3.jpg')
imatge0= imatge[:, :, 1]
plt.imshow(imatge0)
```

```python
imatge= plt.imread ('img3.jpg')
imatge0= imatge[:, :, 2]
plt.imshow(imatge0)
```

Aplicant un colormap podem modificar la fotografia i convertir-la en escala de grisos, o aplicar-hi un altre efecte de color. 

```python
plt.imshow(imatge0, cmap='gray')
```

```python
plt.imshow(imatge0, cmap='twilight')
```

ara, finalment, guardarem les imatges, al directori on estem treballant, amb la funció plt.imsave()

```python
plt.imsave('imatgeGreyscale.jpg',imatge0, cmap='gray')
plt.imsave('imatgeTwilight.jpg',imatge0, cmap='twilight')
```
